# Unsupervised autoencoder pre-training

`learn = create_autoencoder(data, model, path)`

## Source  domain dataset  - i.e. ADNI 

```python
data_source = get_adni('ADNI',num_samples=12, test_size=0.2, transforms=tfms, balance=True,
                       z_factor=0.4)
```

##  Initialize a model

```python
model = ConvAutoEncoder(data_source.shape, act_fn=nn.ELU, debug=True)
model.debug=False
```

## Create a Learner: object connecting a DataBunch with a model

* training/validation/testing loops
* visualizations: reconstruction, training curves, latent space, bottleneck, 
* model/encoder saving: to continue learning later or to transfer encoder weights
* predict/encode test data

```python
device = torch.device("cuda", 0)


learn = Learner(data_source, model, path='ADNI/pretraining', model_type='autoencoder',
                loss_fn=nn.MSELoss(), opt_fn=optim.Adam, device=device, bs=2)
```

## Let's train with test set as validation, and see the reconstructions in every 2nd epoch!

```python
learn.fit(5, lr=2e-3, wd=0.0, l1_reg=1e-7, 
          with_valid=True, show_every=2)
```

```python
learn.plot_history()
```

```python
learn.fit(1, lr=1e-4)
```

## Visualize latent space /w test data

```python
learn.plot_TSNE(data='test', title='tSNE of test data in latent space')
```

```python
learn.plot_bottleneck_histogram(data='test', title='bottleneck activation histogram')
```

## Save full model and encoder to path

```python
learn.save_encoder()
learn.save()
```

# Discriminative fine-tuning: ADNI to PPMI T1 

`learn = create_classifier_from_encoder(data, encoder_path, path)`:


## Target dataset - Parkinson's T12

```python
data_target = get_ppmi('PPMI', mri_type='T1', num_samples=14, balance=True, z_factor=0.75)
```

## Create a Learner for transfer learning

* replaces the default convolutional layer with the pretrained encoder
* freezes the encoder
* returns a classifier Learner

```python
learn = create_classifier_from_encoder(data_target, encoder_path="ADNI/pretraining/models/encoder.pth", 
                                       dropout1=0.05, device=device)
```

```python
learn.model
```

## Let's fine-tune it on the target task (i.e. PD vs. HC)

```python
learn.fit(5, lr=5e-4, with_valid=True)
```

```python
learn.plot_history()
```

```python
preds, cm = learn.predict()
```

## Let's unfreeze the encoder and train the whole stack

```python
learn.unfreeze_encoder()
```

```python
learn.fit(3, lr=1e-3)
```

```python
learn.plot_history()
```
